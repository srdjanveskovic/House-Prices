# Modelling

```python
from sklearn.linear_model import ElasticNet, Lasso
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.kernel_ridge import KernelRidge
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.metrics import mean_squared_error
import xgboost as xgb
import lightgbm as lgb
```

```python
def rmse_cv(model):
    kf = KFold(5, shuffle=True,random_state=42).get_n_splits(train.values)
    rmse = np.sqrt(-cross_val_score(model, train.values, y_train, scoring='neg_mean_squared_error', cv = kf))
    return (rmse)
```

```python
# Lasso Regression
# This model can be sensitive to outliers. That's why we use Robustscaler()
lasso = make_pipeline(RobustScaler(),Lasso(alpha=0.0005,random_state=1))
```

```python
# Elastic Net Regression, an alternative to Lasso
ENet= make_pipeline(RobustScaler(),ElasticNet(alpha=0.0005, l1_ratio=0.9, random_state=3))
```

```python
KRR = KernelRidge(alpha=0.6,kernel='polynomial',degree=2,coef0=2.5)
```

```python
GBoost = GradientBoostingRegressor(n_estimators=1000, learning_rate=0.05, random_state=5)
```

```python
model_xgb = xgb.XGBRegressor(n_estimators=1000, learning_rate=0.05, random_state=7)
```

```python
model_lgb = lgb.LGBMRegressor(objective='regression', learning_rate=0.05, n_estimators=1000)
```

```python
print(f'Lasso score: {round(rmse_cv(lasso).mean(),4)} ({round(rmse_cv(lasso).std(), 4)})')
```

```python
print(f'ElasticNet score: {round(rmse_cv(ENet).mean(),4)} ({round(rmse_cv(ENet).std(), 4)})')
```

```python
print(f'Kernel Ridge score: {round(rmse_cv(KRR).mean(),4)} ({round(rmse_cv(KRR).std(), 4)})')
```

```python
print(f'Gradient Boosting score: {round(rmse_cv(GBoost).mean(),4)} ({round(rmse_cv(GBoost).std(), 4)})')
```

```python
print(f'Xgboost score: {round(rmse_cv(model_xgb).mean(),4)} ({round(rmse_cv(model_xgb).std(), 4)})')
```

```python
print(f'Lgb score: {round(rmse_cv(model_lgb).mean(),4)} ({round(rmse_cv(model_lgb).std(), 4)})')
```

```python
X_train, X_val, y_train, y_val = train_test_split(train, y_train, test_size=0.3, random_state=42)
```

```python
model_xgb.fit(X_train,y_train)
xgb_tr_pred = model_xgb.predict(X_val)
xgb_test_pred = np.exp(model_xgb.predict(test))
```

```python
print(np.sqrt(mean_squared_error(y_val, xgb_tr_pred)))
```

```python
lasso.fit(X_train,y_train)
lasso_tr_pred = lasso.predict(X_val)
laso_test_pred = np.exp(lasso.predict(test))
print(np.sqrt(mean_squared_error(y_val, lasso_tr_pred)))
```

```python
ENet.fit(X_train,y_train)
ENet_tr_pred = ENet.predict(X_val)
ENet_test_pred = np.exp(ENet.predict(test))
print(np.sqrt(mean_squared_error(y_val, ENet_tr_pred)))
```

```python
# Since ENet has a best score, we will use it as our model for prediction.
submission = pd.DataFrame({'Id': test_id, 'SalePrice': ENet_test_pred})
```
